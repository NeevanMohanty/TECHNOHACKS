```python
import time

def slow_print(text):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(0.02)
    print()


def start_game():
    slow_print("\n🔥 Welcome to the Clover Kingdom RPG!")
    name = input("Enter your name, warrior: ")

    slow_print(f"\n{name}, you were born without magic...")
    slow_print("But you dream of becoming the Wizard King!")

    first_choice(name)


def first_choice(name):
    slow_print("\n🌲 You enter a dark forest to train.")
    choice = input("Do you (1) Fight a beast or (2) Run away? ")

    if choice == "1":
        slow_print("\n💥 You defeated the beast and gained strength!")
        exam_scene(name, strong=True)
    elif choice == "2":
        slow_print("\n😨 You ran away. Your confidence drops.")
        exam_scene(name, strong=False)
    else:
        slow_print("Invalid choice!")
        first_choice(name)


def exam_scene(name, strong):
    slow_print("\n🏰 Magic Knight Entrance Exam begins!")

    if strong:
        slow_print("🔥 Your determination impresses everyone!")
        choice = input("Do you (1) Show off power or (2) Stay humble? ")

        if choice == "1":
            slow_print("\n⚡ You attract attention... but also rivals.")
            final_battle(name, power=True)
        else:
            slow_print("\n✨ Your humility earns respect.")
            final_battle(name, power=True)

    else:
        slow_print("😔 You struggle in the exam.")
        choice = input("Do you (1) Keep trying or (2) Give up? ")

        if choice == "1":
            slow_print("\n💪 Your persistence is noticed!")
            final_battle(name, power=False)
        else:
            bad_ending(name)


def final_battle(name, power):
    slow_print("\n⚔️ A powerful enemy appears!")

    choice = input("Do you (1) Fight or (2) Escape? ")

    if choice == "1":
        if power:
            good_ending(name)
        else:
            slow_print("\nYou fight bravely but lack strength...")
            neutral_ending(name)
    else:
        bad_ending(name)


def good_ending(name):
    slow_print(f"\n🏆 {name}, you defeated the enemy!")
    slow_print("You are now a Magic Knight!")
    slow_print("✨ Ending: HERO OF THE CLOVER KINGDOM")


def neutral_ending(name):
    slow_print(f"\n😐 {name}, you survived but failed to win.")
    slow_print("You vow to return stronger.")
    slow_print("⚖️ Ending: UNFINISHED JOURNEY")


def bad_ending(name):
    slow_print(f"\n💀 {name}, your journey ends here...")
    slow_print("You gave up on your dream.")
    slow_print("❌ Ending: FALLEN WARRIOR")


if __name__ == "__main__":
    start_game()
```
